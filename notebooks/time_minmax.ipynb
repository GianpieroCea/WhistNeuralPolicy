```python
from timeit import default_timer
from src.whist.game import WhistGame
from src.whist.minmax import minmax
import numpy as np


def test_alphabetaminmax2(whist : WhistGame, size :int) -> None:

    whist.deal(size=size)
 
    while not whist.is_finished:
        val,card = minmax(whist,26,-np.inf,np.inf,0) 
        print(f"The cards chosen by player {whist.current_player} is {card}")
        whist = WhistGame.apply_move(whist,card)

    print(val)

start = default_timer()
whist = WhistGame()
test_alphabetaminmax2(whist=whist,size=6)

end = default_timer()

print(end-start)
```
